```r
library(tidyverse)
```

```r
set.seed(42)
```

```r
community_data <- matrix(runif(50), nrow=10) # Community data matrix (10 samples x 5 species)
rownames(community_data) <- paste0("Sample", 1:10)
colnames(community_data) <- paste0("Species", 1:5)
community_data
```

```r
environmental_data <- data.frame(Group = rep(c("A", "B"), each=5)) # Grouping variable
environmental_data
```

```r
# Calculate the distance matrix
distance_matrix <- vegdist(community_data, method = "bray")
distance_matrix
```

```r
# Run ADONIS2
adonis_result <- adonis2(distance_matrix ~ Group, data = environmental_data)
print(adonis_result)
```

```r
metadata = read.csv("permanova_metadata.csv", header=TRUE, row.names=1)
head(metadata)
```

```r
transformed = read.csv("permanova_clr_transformed.csv", header=TRUE, row.names=1)
head(transformed)
```

```r
for (timepoint in c(0, 1, 6, 12, 18, 24)) {
  print(timepoint)

  metadata_curr = metadata %>% filter(time_cat == timepoint)
  transformed_curr = transformed[rownames(metadata_curr), ]

  groups = metadata_curr[c("matabx_expo_cat", "del_method")]

  distance_matrix <- vegdist(transformed_curr, method = "euclidean")

  adonis_result <- adonis2(distance_matrix ~ matabx_expo_cat, data = groups)
  print(adonis_result)

}
```
